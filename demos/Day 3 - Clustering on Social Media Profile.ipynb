```python
import pandas as pd
import numpy as np
from sklearn import preprocessing, cluster
```

```python
sns = pd.read_csv("https://raw.githubusercontent.com/abulbasar/data/master/snsdata.csv")
```

```python
sns.head()
```

```python
X = sns.iloc[:, 4:]
X.head()
```

```python
scaler = preprocessing.StandardScaler()
X_std = scaler.fit_transform(X) # Calculates z_score

# Z_score = (x - mean(x))/std(x), for every feature x.
# Goal of Z-score is to get mean = 0 and std = 1 for every feature

pd.DataFrame(X_std).describe()
```

```python
X_std.shape
```

```python
kmeans = cluster.KMeans(n_clusters=5, random_state=1)
y_pred = kmeans.fit_predict(X_std)
```

```python
y_pred
```

```python
kmeans.cluster_centers_.shape
```

```python
clusters = pd.DataFrame(kmeans.cluster_centers_)
clusters
```

```python
clusters["members"] = pd.Series(y_pred).value_counts().sort_index()
clusters.head()
```

```python
X_std[y_pred == 0].mean(axis = 0).shape
```

```python
group_0 = pd.Series(scaler.inverse_transform(kmeans.cluster_centers_[0]))
group_0.index = X.columns
group_0.sort_values(ascending = False).iloc[:10]
```

```python
group_4 = pd.Series(scaler.inverse_transform(kmeans.cluster_centers_[4]))
group_4.index = X.columns
group_4.sort_values(ascending = False).iloc[:10]
```

```python
group_1 = pd.Series(scaler.inverse_transform(kmeans.cluster_centers_[1]))
group_1.index = X.columns
group_1.sort_values(ascending = False).iloc[:10]
```

```python
sns["cluster"] = y_pred
```

```python
sns.groupby(["cluster", "gender"]).gradyear.count().unstack()
```

```python
kmeans.inertia_
```

```python
clusters["avg_distance"] = np.zeros([5])
for i in range(5):
    gr = X_std[y_pred == i]
    center = kmeans.cluster_centers_[i]
    distances = np.sqrt(np.sum((gr - center) ** 2, axis = 1)) # Euclean distance
    print("cluster", i, ":", np.mean(distances))
    clusters.loc[i, "avg_distance"] = np.mean(distances)
clusters.head()
```
