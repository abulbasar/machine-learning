```python
import pandas as pd
pd.options.display.max_columns = 1000
```

```python
df = pd.read_csv("/data/credit-default.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df_dummy = pd.get_dummies(df, drop_first=True)
del df_dummy["default"]
```

```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.StandardScaler() # z_score = (x - mean(x))/std(x)
df_std = scaler.fit_transform(df_dummy)
# purpose is make 0 mean and 1 stddev
```

```python
pd.DataFrame(df_std).describe()
```

```python
X = df_std
y = df.default
```

```python
from sklearn import model_selection
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y
                        , test_size = 0.3, random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_test_pred = lr.predict(X_test)
```

```python
result = pd.DataFrame({"actual": y_test, "prediction": y_test_pred})
result["match"] = result.actual == result.prediction
result
```

```python
result[result.match].shape
```

```python
accuracy = 227/len(X_test)
accuracy 
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test, y_test_pred)
```
