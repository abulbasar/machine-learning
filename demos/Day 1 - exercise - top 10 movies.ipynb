```python
import pandas as pd
```

```python
movies = pd.read_csv("/data/movielens/movies.csv")
ratings = pd.read_csv("/data/movielens/ratings.csv")
```

```python
movies.shape[0]
```

```python
ratings.shape[0]
```

```python
movies[movies.title.str.contains("Godfather")]
```

```python
god_father_movie_id = int(movies[movies.title == "Godfather, The (1972)"].movieId)
god_father_movie_id
```

```python
ratings[ratings.movieId == god_father_movie_id].shape
```

```python
rating_avg = ratings.groupby("movieId").rating.agg(["mean", "count"])
rating_avg
```

```python
rating_avg.loc[god_father_movie_id, :]["mean"]
```

```python
top10 = rating_avg[rating_avg["count"]>=100]\
.sort_values("mean", ascending = False).iloc[:10,:]

top10
```

```python
top10.merge(movies, on = "movieId")
```
