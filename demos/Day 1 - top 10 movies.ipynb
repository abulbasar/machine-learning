```python
rating_avg = ratings.groupby("movieId").rating.agg(["count","mean"])
rating_avg
```

```python
rating_avg_top_100 = rating_avg[rating_avg["count"]>100]
rating_avg_top_100.shape
```

```python
top_10 = rating_avg_top_100.sort_values("mean", ascending = False).iloc[:10, :]
```
