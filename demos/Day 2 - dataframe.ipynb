```python
iqr = q3 - q1
```

```python
def outliers(y):
    q1, q2, q3 = np.percentile(y, [25, 50, 75])
    iqr = q3 - q1
    uw = min(np.max(y), q3 + 1.5 * iqr)
    lw = max(np.min(y), q1 - 1.5 * iqr)
    print("uw:", uw, "lw: ", lw)
    return np.where((y > uw) | (y < lw), True, False)
```

```python
df[outliers(df.charges)].sort_values("charges")
```
