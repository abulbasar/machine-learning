```python
import pandas as pd
from sklearn import * 
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/abulbasar/data/master/kaggle-houseprice/data_combined_cleaned.csv")
```

```python
df.head()
```

```python
del df["Id"]
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
df.info()
```

```python
target = "SalesPrice"

y = np.log(df[target])
X = df.drop(columns=target)
X_dummy = pd.get_dummies(X, drop_first=True)

features = X_dummy.columns

X_train, X_test, y_train, y_test = model_selection.train_test_split(X_dummy, y
                , test_size = 0.3, random_state = 1)


pipe = pipeline.Pipeline([
    ("poly", preprocessing.PolynomialFeatures(degree = 1
                                    , include_bias=False)),
    ("scaler", preprocessing.StandardScaler()),
    ("est", linear_model.LinearRegression())
])

pipe.fit(X_train, y_train)

y_test_pred = pipe.predict(X_test)
y_train_pred = pipe.predict(X_train)


print("train r2", metrics.r2_score(y_train, y_train_pred)) 
print("test r2", metrics.r2_score(y_test, y_test_pred))

print("train rmse", metrics.mean_squared_error(y_train, y_train_pred) ** 0.5) 
print("test rmse", metrics.mean_squared_error(y_test, y_test_pred) ** 0.5)

#pd.DataFrame({"feature": features, "coeff": est.coef_})
```

```python
summary = pd.DataFrame({"y_pred": y_test_pred, "y": y_test})
```

```python
summary.sort_values("y_pred", ascending=False)
```

```python
est = pipe.steps[-1][-1]
coeff = pd.DataFrame({"f": features, "w": est.coef_})
coeff.sort_values("w", ascending = False)
```

```python
target = "SalesPrice"

y = np.log(df[target])
X = df.drop(columns=target)
X_dummy = pd.get_dummies(X, drop_first=True)

features = X_dummy.columns

X_train, X_test, y_train, y_test = model_selection.train_test_split(X_dummy, y
                , test_size = 0.3, random_state = 1234)


pipe = pipeline.Pipeline([
    ("poly", preprocessing.PolynomialFeatures(degree = 1
                                    , include_bias=False)),
    ("scaler", preprocessing.StandardScaler()),
    ("est", linear_model.Lasso(alpha = 0.01, random_state = 1))
])

pipe.fit(X_train, y_train)

y_test_pred = pipe.predict(X_test)
y_train_pred = pipe.predict(X_train)


print("train r2", metrics.r2_score(y_train, y_train_pred)) 
print("test r2", metrics.r2_score(y_test, y_test_pred))

print("train rmse", metrics.mean_squared_error(y_train, y_train_pred) ** 0.5) 
print("test rmse", metrics.mean_squared_error(y_test, y_test_pred) ** 0.5)

#pd.DataFrame({"feature": features, "coeff": est.coef_})
```

```python
est = pipe.steps[-1][-1]
coeff = pd.DataFrame({"f": features, "w": est.coef_})
coeff.sort_values("w", ascending = False)
```

```python
%%time 

pipe = pipeline.Pipeline([
    ("poly", preprocessing.PolynomialFeatures(degree = 1
                                    , include_bias=False)),
    ("scaler", preprocessing.StandardScaler()),
    ("est", linear_model.ElasticNet(random_state = 1))
])
param_grid = {
    "est__alpha": (np.random.random(10) - 0.5) * 0.01 + 0.1,
    "est__l1_ratio": np.random.random(10)
}

gsearch = model_selection.GridSearchCV(pipe, cv = 5, n_jobs=8, verbose=True
                        , param_grid = param_grid)


gsearch.fit(X_dummy, y)
```
