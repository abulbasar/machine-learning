```python
stocks["pct_change"] = (stocks.close - stocks.open) * 100 / stocks.open
stocks[["symbol", "date", "open", "close", "pct_change"]]
```

```python
stocks.sort_values("pct_change"
    , ascending = False)[["symbol", "date", "pct_change"]].iloc[:10, :]
```
