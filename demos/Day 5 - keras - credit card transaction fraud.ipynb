```python
import keras
import numpy as np

import pandas as pd

pd.options.display.max_columns = 1000
pd.options.display.max_rows = 10

```

```python
df = pd.read_csv("/data/creditcard-fraud.csv")
df
```

```python
training_size = int(len(df) * 0.7) 
```

```python
X_train = df.iloc[:training_size, 1:-2].values
y_train = df.iloc[:training_size, -1]
X_test = df.iloc[training_size:, 1:-2].values
y_test = df.iloc[training_size:, -1]

X_mean, X_std = X_train.mean(axis=0), X_train.std(axis=0)

X_train = (X_train - X_mean)/X_std
X_test = (X_test - X_mean)/X_std
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn import linear_model, metrics
```

```python
est = linear_model.SGDClassifier(max_iter=1000)
est.fit(X_train, y_train)
y_test_pred = est.predict(X_test)

metrics.confusion_matrix(y_test, y_test_pred)
```

```python
n_x = X_train.shape[1]
```

```python
from keras.layers import InputLayer, Dense, Dropout
```

```python
model = keras.Sequential()
model.add(InputLayer((n_x, )))
model.add(Dense(10, activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(10, activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(10, activation="relu"))
model.add(Dense(1, activation="sigmoid"))

model.compile("adam", keras.losses.binary_crossentropy, ["accuracy"])

model.summary()
```

```python
model.fit(X_train, y_train, batch_size=128, validation_data=(X_test, y_test), epochs=10)
```

```python
y_test_pred = model.predict_classes(X_test)
metrics.confusion_matrix(y_test, y_test_pred)
```
