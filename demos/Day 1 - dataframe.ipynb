```python
df.groupby("gender").charges.mean()
```

```python
df.groupby("gender").charges.agg(["count", "mean"])
```

```python
df.groupby("gender").agg({"charges": ["count", "mean"]
                    , "age": ["mean", "median"]})
```

```python
df.groupby(["region", "gender"]).agg({"charges": ["count", "mean"]
                    , "age": ["mean", "median"]})
```

```python
df.pivot_table("charges", "region", "gender", aggfunc="median")
```
