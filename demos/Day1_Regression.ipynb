```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/abulbasar/data/master/insurance.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
X = df.iloc[:, :-1]
X.head()
```

```python
y = df.charges
```

```python
X_dummy = pd.get_dummies(X, drop_first=True)
X_dummy.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_std = scaler.fit_transform(X_dummy)
```

```python
type(X_std)
```

```python
pd.DataFrame(X_std)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_std, y,
                                                    test_size = 0.3, random_state = 1)
```

```python
import numpy as np
```

```python
np.random.seed(10000)
np.random.rand(3)
```

```python
X_train.shape[0] / X.shape[0]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression(normalize=False)
lr.fit(X_train, y_train)
```

```python
y_test_pred = lr.predict(X_test)
```

```python
output = pd.DataFrame({"actual": y_test, "predicted": y_test_pred})
output["error"] = (output.actual - output.predicted) / output.actual
output
```

```python
from sklearn import metrics
```

```python
metrics.mean_squared_error(y_test, y_test_pred)
```

```python
metrics.r2_score(y_test, y_test_pred)
```
