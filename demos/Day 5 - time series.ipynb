```python
import numpy as np
```

```python
aapl_close = np.log(aapl[["adjclose"]])
aapl_close
```

```python
aapl_close.resample("Q").agg(["min", "max"])
```

```python
aapl_close.resample("W").agg(["min", "max"])
```

```python

aapl_close["3d_mean"] = aapl_close.rolling(3).mean()
aapl_close["3d_mean_1lag"] = aapl_close["3d_mean"].shift(1)
aapl_close
```

```python
est = aapl_close.dropna()
```

```python
from sklearn import *
```

```python
metrics.r2_score(est.adjclose, est["3d_mean_1lag"])
```

```python
metrics.mean_squared_error(est.adjclose, est["3d_mean_1lag"]) ** 0.5
```
