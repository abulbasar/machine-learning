```python
class NegativeAmountException(Exception):pass
class InsufficientBalanceException(Exception):pass
```

```python
class Account:
    """
    Account class hold information about account of a person.
    It has methods for account number generation, withdrawal and deposit
    """
    
    account_ids = []
    
    @staticmethod
    def generate_id():
        return len(Account.account_ids) + 1
    
    def __init__(self, name, amount = 0.0):
        self.acc_id = Account.generate_id()
        Account.account_ids.append(self.acc_id)
        self.name = name
        self.amount = amount
        
    def __repr__(self):
        return "[Account] name: " + self.name + " amount: " \
                    + str(self.amount) + " acc id:  " + str(self.acc_id)
    
    
    def deposit(self, amount):
        """
            Withdraw amount from the account object. 
            @Input amount: float - amount to withdraw
            
            @returns
            None
        """
        self.amount = self.amount + amount
        
    def withdraw(self, amount:float):
        
        if amount < 0:
            raise NegativeAmountException("amount must be positive")
        if amount > self.amount:
            raise InsufficientBalanceException("In sufficient balance in your account")
        
        self.amount = self.amount - amount

x = Account("Bill Gates")
x.deposit(100)
try:
    x.withdraw("30")
    
except NegativeAmountException as e:
    print("Negative amount exception during transaction - ", e)
    
except InsufficientBalanceException as e:
    print("Insufficient balance exception during transaction - ", e)
    
except Exception as e:
    print("Unknown exception during transaction - ", e)
    
x
```
