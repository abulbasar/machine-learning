````python
# q1: it gives a value, x for a variable (y), for which 25% of the population falls below the value x,
# it also implies that 75% of the population is above the value x, for the variable y.
````

````python
q1, q3 = np.percentile(df.price, [0.25, 0.75])
q1, q3
````

````python
iqr = q3 - q1
iqr
````

````python
lower_whisker = max(df.price.min(), q1 - 1.5 * iqr)
upper_whisker = min(df.price.max(), q3 + 1.5 * iqr)

lower_whisker, upper_whisker
````

````python
def is_outlier(a):
    q1, q3 = np.percentile(a, [25, 75])
    print(q1, q3)
    iqr = q3 - q1
    lower_whisker = max(a.min(), q1 - 1.5 * iqr)
    upper_whisker = min(a.max(), q3 + 1.5 * iqr)
    print("lower whisker:", lower_whisker, "upper whisker:", upper_whisker)
    return (a < lower_whisker) | (a > upper_whisker)

is_outlier(df.price)
````
