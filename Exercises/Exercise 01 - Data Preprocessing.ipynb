### LotFrontage: Linear feet of street connected to property
There are 486 missing values for LotFrontage, which is quite a lot of values to fill and we canâ€™t just replace these with 0. Weâ€™re given that â€œLotFrontage: Linear feet of street connected to property.â€ The area of each street connected to the house property is most likely going to have a similar area to other houses in its neighborhood. We can group by each neighborhood and take the median of each LotFrontage and fill the missing values of each LotFrontage based on what neighborhood the house comes from.



```python
lot_columns = find_columns(df, "lot")
lot_columns.append("Neighborhood")
lot_columns
```

```python
neighborhood = df.groupby("Neighborhood").LotFrontage.agg([len, np.median])
neighborhood
```

```python
df2 = df[lot_columns].merge(neighborhood, 
                      left_on="Neighborhood", right_index=True)
df2
```

```python
idx = df[df.LotFrontage.isnull()].index
df.loc[idx, "LotFrontage"] = df2.loc[idx, "median"]
df[lot_columns].merge(neighborhood, 
                      left_on="Neighborhood", right_index=True)
```

```python
find_null_columns()
```
